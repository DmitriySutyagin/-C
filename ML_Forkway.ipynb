```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english') + stopwords.words('russian'))

# List it skills

it_terms = ["python", "javascript", "c++", "java", "swift", "kotlin", "go", "rust", "django", "flask", "react", "angular", "spring", "hibernate", "sql", "postgresql", "mysql", "mongodb", "redis", "git", "svn",
            "linux", "windows", "macos", "unit-тесты", "интеграция", "автоматизация", "jenkins", "travisci", "circleci", "docker", "kubernetes", "шифрование", "аутентификация", "авторизация", "html/css",
            "typescript", "bootstrap", "tailwind css", "node.js", "express.js", "ruby on rails", "aws", "google cloud", "azure", "restful", "graphql", "ansible", "terraform", "pandas", "numpy", "matplotlib",
            "seaborn", "scikit-learn", "tensorflow", "pytorch", "keras", "amazon sagemaker", "google ai platform", "tcp/ip", "dns", "dhcp", "apache", "nginx", "prometheus", "grafana", "puppet", "chef",
            "разработка мобильных приложений", "работа с большими данными", "облачные вычисления", "кибербезопасность", "сетевое администрирование", "проектирование баз данных", "управление проектами",
            "agile-методологии", "devsecops", "виртуализация", "кластеризация", "отказоустойчивость", "машинное обучение", "обработка естественного языка", "компьютерное зрение", "распределённые системы",
            "микросервисная архитектура", "серверная оптимизация", "мониторинг производительности", "управление качеством по", "тестирование безопасности", "ux/ui дизайн", "seo оптимизация", "аналитика данных",
            "визуализация данных", "алгоритмы и структуры данных", "криптография", "высоконагруженные системы", "параллельное программирование", "интернет вещей (iot)", "блокчейн", "квантовые вычисления", "защита от ddos атак", "работа с nosql базами данных", "нагрузочное тестирование", "кросс-платформенная разработка", "реактивное программирование", "разработка игр", "создание чат-ботов", "нативное мобильное приложение", "программирование микроконтроллеров", "обработка изображений", "распознавание речи", "системы рекомендаций", "бизнес-аналитика", "информационная безопасность", "системы управления знаниями", "проектировка интерфейсов", "разработка высокопроизводительных систем", "нейронные сети", "глубокое обучение", "статистический анализ", "работа с api", "мониторинг инфраструктуры", "анализ логов", "техническое документирование", "работа с требованиями", "техподдержка", "развёртывание решений", "работа с финансовыми системами", "работа с erp-системами", "прогнозная аналитика", "ar/vr разработка", "цифровое производство", "моделирование процессов", "управление рисками", "сопровождение проектов"]

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]
    filtered_tokens = [lemma for lemma in lemmas if lemma not in stop_words]

     # Оставляем только IT-термины
    it_skills = [token for token in filtered_tokens if token in it_terms]

    return it_skills
```

```python
skills_dict = {}

for index, row in data.iterrows():
    skills = preprocess_text(row['description'])
    for skill in skills:
        if skill not in skills_dict:
            skills_dict[skill] = []
        skills_dict[skill].append(str(row['id']))
```
